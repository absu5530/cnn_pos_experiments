## Data split

```python
def split_data(training_essays,set_no):
    df_set = training_essays[training_essays['essay_set'] == set_no] #get particular essay set
    df_set_subset = df_set[['essay_id','essay_set' ,'essay','rater1_domain1','rater2_domain1']] #get only desired columns
    df_set_subset = df_set_subset.dropna() #drop rows with na values

    for couple in df_set_subset['rater1_domain1'].value_counts().iteritems(): #ensure that every score point has at least 4 samples
        if couple[1] < 4:
            print(couple[0])
            df_set_subset = df_set_subset[df_set_subset['rater1_domain1'] != couple[0]]
    
    Data = df_set_subset[['essay_id','essay_set','essay','rater1_domain1','rater2_domain1']]
    Target = df_set_subset[['rater1_domain1']]

    X_majority, X_test, y_majority, y_test = train_test_split(Data, Target, test_size=0.2, random_state=42, stratify = Target)
    y_test['set'] = 'test'
    X_train, X_dev, y_train, y_dev = train_test_split(X_majority, y_majority, test_size=0.25, random_state=42, stratify = y_majority)
    y_train['set'] = 'train'
    y_dev['set'] = 'dev'
    
    return X_train, y_train, X_dev, y_dev, X_test, y_test
```

## Prompt-wise statistics

```python
def prompt_tables(train_data,dev_data,test_data):    
    
    def split_details(data, name):
        disagreements = 0
        total = 0
        for index, sample in data.iterrows():
            total += 1
            if sample['rater1_domain1'] != sample['rater2_domain1']:
                disagreements += 1
        df = pd.DataFrame(data['rater1_domain1'].value_counts().sort_index()).rename(columns={'rater1_domain1': name}).transpose()
        df.columns = ['Score point ' + str(col) for col in df.columns]
        df['Total'] = total
        df['Disagreement rate'] = (disagreements*100)/total
        return(df)
    
    dict_words = {}
    wc = []
    c = 0
    for record in train_data.append(dev_data).append(test_data)['essay']:
        wc.append(len(word_tokenize(record)))
        c += 1
        for word in word_tokenize(record):
            if word not in dict_words:
                dict_words[word] = 1
        
    
    avg_wc = np.mean(wc)
    max_wc = np.max(wc)
    min_wc = np.min(wc)
    std_wc = np.std(wc)
    
    stats_table = split_details(train_data,'Train').append(split_details(dev_data, 'Dev')).append(split_details(test_data, 'Test'))
    return avg_wc,std_wc,c,len(dict_words.keys()),max_wc,min_wc, stats_table
```
